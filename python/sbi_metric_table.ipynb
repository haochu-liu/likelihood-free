```python
import numpy as np
import pandas as pd
import itertools
```

Use five metrics for estimated posterior distribution
- coverage of 95\% confidence interval
- bias of posterior mean
- posterior variance
- relative bias
- relative root mean square error

```python
list_1 = ['1D_exact', '1D_prop', '1D_all', '1D_long',
          '1D_summaryNN', '1D_lr', '1D_GAM', '1D_predNN',
          '2D_exact', '2D_prop', '2D_all', '2D_long']
list_2 = ['SimBac500', 'SimBac2000', 'SimBac6000',
          'ClonalOrigin500', 'ClonalOrigin2000', 'ClonalOrigin6000']

combinations = list(itertools.product(list_1, list_2))
df = pd.DataFrame(combinations, columns=['Method', 'Observations'])

df['rho_coverage'] = None
df['rho_bias'] = None
df['rho_variance'] = None
df['rho_relative_bias'] = None
df['rho_relative_RMSE'] = None

df['theta_coverage'] = None
df['theta_bias'] = None
df['theta_variance'] = None
df['theta_relative_bias'] = None
df['theta_relative_RMSE'] = None

df['L_coverage'] = None
df['L_bias'] = None
df['L_variance'] = None
df['L_relative_bias'] = None
df['L_relative_RMSE'] = None

df
```

```python
df.to_csv('../output/sbi_metric_table.csv', index=False)
```
