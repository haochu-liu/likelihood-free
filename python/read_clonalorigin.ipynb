## Simulated observations

```python
rho_fix = 0.02
theta_fix = 0.05
delta_fix = 300
L = [500, 2000, 6000]

np.random.seed(1000)
```

```python
x_o_500  = np.full((100, 46), np.nan)
x_o_2000 = np.full((100, 46), np.nan)
x_o_6000 = np.full((100, 46), np.nan)
x_o_500.shape, x_o_2000.shape, x_o_6000.shape
```

```python
for i in tqdm(range(100), desc="Simulating ClonalOrigin sequences of length 500"):
    s_vec = ClonalOrigin_seq_sim(clonal_tree, rho_fix, theta_fix, L[0], delta_fix)
    x_o_500[i, :] = s_vec
```

```python
for i in tqdm(range(100), desc="Simulating ClonalOrigin sequences of length 2000"):
    s_vec = ClonalOrigin_seq_sim(clonal_tree, rho_fix, theta_fix, L[1], delta_fix)
    x_o_2000[i, :] = s_vec
```

```python
for i in tqdm(range(100), desc="Simulating ClonalOrigin sequences of length 6000"):
    s_vec = ClonalOrigin_seq_sim(clonal_tree, rho_fix, theta_fix, L[2], delta_fix)
    x_o_6000[i, :] = s_vec
```

```python
x_o_500, x_o_2000, x_o_6000
```

```python
np.savetxt("../data/ClonalOrigin/x_o_500.csv", x_o_500, delimiter=",")
np.savetxt("../data/ClonalOrigin/x_o_2000.csv", x_o_2000, delimiter=",")
np.savetxt("../data/ClonalOrigin/x_o_6000.csv", x_o_6000, delimiter=",")
```
