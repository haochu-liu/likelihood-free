# Plots for the benchmarking test

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## Load data

```python
model_names = ['NLE', 'NPE', 'NRE']
sim_budgets = ['200', '500', '1000', '2000', '5000', '10000', '50000']
```

```python
nle_marginal_coverage = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nle_marginal_coverage_results{i}.npy'
    nle_marginal_coverage_i = np.load(load_path)
    nle_marginal_coverage[:, :, i-1] = nle_marginal_coverage_i[:, :, 0]

npe_marginal_coverage = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/npe_marginal_coverage_results{i}.npy'
    npe_marginal_coverage_i = np.load(load_path)
    npe_marginal_coverage[:, :, i-1] = npe_marginal_coverage_i[:, :, 0]

nre_marginal_coverage = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nre_marginal_coverage_results{i}.npy'
    nre_marginal_coverage_i = np.load(load_path)
    nre_marginal_coverage[:, :, i-1] = nre_marginal_coverage_i[:, :, 0]

marginal_coverage = [nle_marginal_coverage, npe_marginal_coverage, nre_marginal_coverage]
```

```python
nle_multidim_coverage = np.full((7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nle_multidim_coverage_results{i}.npy'
    nle_multidim_coverage_i = np.load(load_path)
    nle_multidim_coverage[:, i-1] = nle_multidim_coverage_i[0, :]

npe_multidim_coverage = np.full((7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/npe_multidim_coverage_results{i}.npy'
    npe_multidim_coverage_i = np.load(load_path)
    npe_multidim_coverage[:, i-1] = npe_multidim_coverage_i[0, :]

nre_multidim_coverage = np.full((7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nre_multidim_coverage_results{i}.npy'
    nre_multidim_coverage_i = np.load(load_path)
    nre_multidim_coverage[:, i-1] = nre_multidim_coverage_i[0, :]

multidim_coverage = [nle_multidim_coverage, npe_multidim_coverage, nre_multidim_coverage]
```

```python
nle_L2_error = np.full((998, 7, 10), np.nan)
nle_mahalanobis_error = np.full((998, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nle_mean_error_results{i}.npy'
    nle_mean_error_i = np.load(load_path)
    nle_L2_error[:, :, i-1] = nle_mean_error_i[:, :, 0]
    nle_mahalanobis_error[:, :, i-1] = nle_mean_error_i[:, :, 2]

npe_L2_error = np.full((998, 7, 10), np.nan)
npe_mahalanobis_error = np.full((998, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/npe_mean_error_results{i}.npy'
    npe_mean_error_i = np.load(load_path)
    npe_L2_error[:, :, i-1] = npe_mean_error_i[:, :, 0]
    npe_mahalanobis_error[:, :, i-1] = npe_mean_error_i[:, :, 2]

nre_L2_error = np.full((998, 7, 10), np.nan)
nre_mahalanobis_error = np.full((998, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nre_mean_error_results{i}.npy'
    nre_mean_error_i = np.load(load_path)
    nre_L2_error[:, :, i-1] = nre_mean_error_i[:, :, 0]
    nre_mahalanobis_error[:, :, i-1] = nre_mean_error_i[:, :, 2]

L2_error = [nle_L2_error, npe_L2_error, nre_L2_error]
mahalanobis_error = [nle_mahalanobis_error, npe_mahalanobis_error, nre_mahalanobis_error]
```

```python
nle_sbc_D = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nle_sbc_D{i}.csv'
    nle_sbc_D_i = np.loadtxt(load_path, delimiter=',')
    nle_sbc_D[:, :, i-1] = nle_sbc_D_i

npe_sbc_D = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/npe_sbc_D{i}.csv'
    npe_sbc_D_i = np.loadtxt(load_path, delimiter=',')
    npe_sbc_D[:, :, i-1] = npe_sbc_D_i

nre_sbc_D = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nre_sbc_D{i}.csv'
    nre_sbc_D_i = np.loadtxt(load_path, delimiter=',')
    nre_sbc_D[:, :, i-1] = nre_sbc_D_i

sbc_D = [nle_sbc_D, npe_sbc_D, nre_sbc_D]
```

```python
nle_sbc_p_values = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nle_sbc_p_values{i}.csv'
    nle_sbc_p_values_i = np.loadtxt(load_path, delimiter=',')
    nle_sbc_p_values[:, :, i-1] = nle_sbc_p_values_i

npe_sbc_p_values = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/npe_sbc_p_values{i}.csv'
    npe_sbc_p_values_i = np.loadtxt(load_path, delimiter=',')
    npe_sbc_p_values[:, :, i-1] = npe_sbc_p_values_i

nre_sbc_p_values = np.full((3, 7, 10), np.nan)
for i in range(1, 11):
    load_path = f'../data/benchmark/nre_sbc_p_values{i}.csv'
    nre_sbc_p_values_i = np.loadtxt(load_path, delimiter=',')
    nre_sbc_p_values[:, :, i-1] = nre_sbc_p_values_i

sbc_p_values = [nle_sbc_p_values, npe_sbc_p_values, nre_sbc_p_values]
```

## Figures

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Coverage', fontsize=14, fontweight='bold', y=1.05)

results = marginal_coverage
x_base = np.arange(len(sim_budgets))
x = np.repeat(x_base, 10)
offsets = [-0.18, 0.0, 0.18]
for i, ax in enumerate(axes):
    y1 = results[i][0, :, :].flatten()
    y2 = results[i][1, :, :].flatten()
    y3 = results[i][2, :, :].flatten()

    ax.scatter(x + offsets[0], y1, marker='x', color='blue', alpha=0.5, label=r'$\rho_s$')
    ax.scatter(x + offsets[1], y2, marker='x', color='green', alpha=0.5, label=r'$\theta_s$')
    ax.scatter(x + offsets[2], y3, marker='x', color='orange', alpha=0.5, label='Gene length')

    ax.hlines(
        y=0.95,
        xmin=x_base[0] - 0.35,
        xmax=x_base[-1] + 0.35,
        color='red',
        linestyle='--',
        linewidth=1,
        label='Target coverage: 95.0%'
    )

    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(x_base)
    ax.set_xticklabels(sim_budgets, rotation=90)
    ax.set_xlim(x_base[0] - 0.5, x_base[-1] + 0.5)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')

    ax.tick_params(axis='y', length=0)
    ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1.0))

axes[0].set_ylabel('Coverage', fontsize=12)
axes[0].legend(loc='lower right', fontsize=10, frameon=False)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.ylim(0, 1)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_coverage.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Parameter Estimation Error (L2 Distance)', fontsize=14, fontweight='bold', y=1.05)

raw_results = L2_error
x_positions = np.arange(len(sim_budgets))

for i, ax in enumerate(axes):
    arr = raw_results[i]
    arr_swapped = np.swapaxes(arr, 1, 2)
    arr_flat = arr_swapped.reshape(-1, 7)

    bplot = ax.boxplot(
        arr_flat,
        positions=x_positions,
        patch_artist=True,
        widths=0.4,
        showfliers=False
    )
    
    for patch in bplot['boxes']:
        patch.set_facecolor('white')
        patch.set_edgecolor('#1f77b4')
        patch.set_linewidth(1.5)
        
    for element in ['whiskers', 'caps', 'medians']:
        for line in bplot[element]:
            line.set_color('#1f77b4')
            line.set_linewidth(1.5)

    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(x_positions)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(axis='y', length=0)

axes[0].set_ylabel(r'Estimation Error', fontsize=12)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_error_box_L2.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
