# ClonalOrigin summary statistics

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Clonal tree:**
* `tree = ClonalTree(n=15)`

**Simulate observed data**

Set the true parameters as
* $\rho_s = 0.02$,
* $\delta = 300$,
* $\theta_s = 0.05$.

**Define prior distribution**

We set the prior to be uniform with range:
* $[0, 0.2]$ for $\rho_s$,
* $[1, 500]$ for $\delta$,
* $[0, 0.2]$ for $\theta_s$.

```python
x_mat = np.loadtxt('../data/ClonalOrigin/x_mat.csv', delimiter=",", skiprows=1)
x_mat.shape
```

```python
theta_site_mat = np.loadtxt('../data/ClonalOrigin/theta_site_mat.csv', delimiter=",", skiprows=1)
rho_site_mat = np.loadtxt('../data/ClonalOrigin/rho_site_mat.csv', delimiter=",", skiprows=1)
delta_mat = np.loadtxt('../data/ClonalOrigin/delta_mat.csv', delimiter=",", skiprows=1)
theta_site_mat.shape, rho_site_mat.shape, delta_mat.shape
```

Fixed $\delta$ and $\theta_s$, only changing $\rho_s$.

```python
fig, axs = plt.subplots(4, 1, figsize=(8, 16))

axs[0].scatter(x_mat[:, 0], rho_site_mat[:, 0], marker='x', color='darkred')
axs[0].scatter(x_mat[:, 0], rho_site_mat[:, 1], marker='x', color='darkblue')
axs[0].scatter(x_mat[:, 0], rho_site_mat[:, 2], marker='x', color='darkgreen')
axs[0].set_ylabel('LD_r')

axs[1].scatter(x_mat[:, 0], rho_site_mat[:, 3], marker='x', color='darkred')
axs[1].scatter(x_mat[:, 0], rho_site_mat[:, 4], marker='x', color='darkblue')
axs[1].scatter(x_mat[:, 0], rho_site_mat[:, 5], marker='x', color='darkgreen')
axs[1].set_ylabel('G3')

axs[2].scatter(x_mat[:, 0], rho_site_mat[:, 6], marker='x', color='darkred')
axs[2].set_ylabel('Homoplasy index')

axs[3].scatter(x_mat[:, 0], rho_site_mat[:, 7], marker='x', color='darkred')
axs[3].set_xlabel('rho per site')
axs[3].set_ylabel('S')

plt.tight_layout()
plt.show()
```

Fixed $\rho_s$ and $\theta_s$, only changing $\delta$.

```python
fig, axs = plt.subplots(4, 1, figsize=(8, 16))

axs[0].scatter(x_mat[:, 1], delta_mat[:, 0], marker='x', color='darkred')
axs[0].scatter(x_mat[:, 1], delta_mat[:, 1], marker='x', color='darkblue')
axs[0].scatter(x_mat[:, 1], delta_mat[:, 2], marker='x', color='darkgreen')
axs[0].set_ylabel('LD_r')

axs[1].scatter(x_mat[:, 1], delta_mat[:, 3], marker='x', color='darkred')
axs[1].scatter(x_mat[:, 1], delta_mat[:, 4], marker='x', color='darkblue')
axs[1].scatter(x_mat[:, 1], delta_mat[:, 5], marker='x', color='darkgreen')
axs[1].set_ylabel('G3')

axs[2].scatter(x_mat[:, 1], delta_mat[:, 6], marker='x', color='darkred')
axs[2].set_ylabel('Homoplasy index')

axs[3].scatter(x_mat[:, 1], delta_mat[:, 7], marker='x', color='darkred')
axs[3].set_xlabel('delta')
axs[3].set_ylabel('S')

plt.tight_layout()
plt.show()
```

Fixed $\rho_s$ and $\delta$, only changing $\theta_s$.

```python
fig, axs = plt.subplots(4, 1, figsize=(8, 16))

axs[0].scatter(x_mat[:, 2], theta_site_mat[:, 0], marker='x', color='darkred')
axs[0].scatter(x_mat[:, 2], theta_site_mat[:, 1], marker='x', color='darkblue')
axs[0].scatter(x_mat[:, 2], theta_site_mat[:, 2], marker='x', color='darkgreen')
axs[0].set_ylabel('LD_r')

axs[1].scatter(x_mat[:, 2], theta_site_mat[:, 3], marker='x', color='darkred')
axs[1].scatter(x_mat[:, 2], theta_site_mat[:, 4], marker='x', color='darkblue')
axs[1].scatter(x_mat[:, 2], theta_site_mat[:, 5], marker='x', color='darkgreen')
axs[1].set_ylabel('G3')

axs[2].scatter(x_mat[:, 2], delta_mat[:, 6], marker='x', color='darkred')
axs[2].set_ylabel('Homoplasy index')

axs[3].scatter(x_mat[:, 2], theta_site_mat[:, 7], marker='x', color='darkred')
axs[3].set_xlabel('theta per site')
axs[3].set_ylabel('S')

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(4, 3, figsize=(16, 18), sharey='row')

x_data = [x_mat[:, 0], x_mat[:, 1], x_mat[:, 2]]
y_data_matrices = [rho_site_mat, delta_mat, theta_site_mat]
x_labels = [r'$\rho_s$', r'$\delta$', r'$\theta_s$']

for col in range(3):
    x = x_data[col]
    y_mat = y_data_matrices[col]

    axs[0, col].scatter(x, y_mat[:, 0]-y_mat[:, 1], marker='x', color='darkred', label= r"$\text{LD}_{50}-\text{LD}_{200}$")
    axs[0, col].scatter(x, y_mat[:, 0]-y_mat[:, 2], marker='x', color='darkgreen', label= r"$\text{LD}_{50}-\text{LD}_{2000}$")

    axs[1, col].scatter(x, y_mat[:, 3], marker='x', color='darkred', label = "50bp")
    axs[1, col].scatter(x, y_mat[:, 4], marker='x', color='darkblue', label = "200bp")
    axs[1, col].scatter(x, y_mat[:, 5], marker='x', color='darkgreen', label = "2000bp")

    axs[2, col].scatter(x, y_mat[:, 6], marker='x', color='darkred')

    axs[3, col].scatter(x, y_mat[:, 7], marker='x', color='darkred')
    axs[3, col].set_xlabel(x_labels[col])

axs[0, 0].set_ylabel(r'Mean $r^2$ difference')
axs[1, 0].set_ylabel('Mean G3')
axs[2, 0].set_ylabel('Homoplasy index')
axs[3, 0].set_ylabel('S')

axs[0, 0].legend(loc='best')
axs[1, 0].legend(loc='best')

plt.tight_layout()
plt.show()
```
