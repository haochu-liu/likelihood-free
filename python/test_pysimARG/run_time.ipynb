```python
ClonalOrigin_pair_time = []
add_mutation_time = []

for i in range(1000):
    np.random.seed(i)

    start_time_1 = time.time()
    ARG = ClonalOrigin_pair.ARG(tree, rho_site, L, delta, k)
    end_time_1 = time.time()
    ClonalOrigin_pair_time.append(end_time_1 - start_time_1)

    start_time_2 = time.time()
    node_site = add_mutation.add_mutation(ARG, theta_site)
    end_time_2 = time.time()
    add_mutation_time.append(end_time_2 - start_time_2)

    if (i+1) % 100 == 0:
        print(f"Complete {i+1} iterations.")
```

```python
np.mean(ClonalOrigin_pair_time), np.mean(add_mutation_time)
```
