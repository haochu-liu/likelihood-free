```python
import numpy as np
import time
import sys
sys.path.append('../../pysimARG')
from clonal_genealogy import ClonalTree
import ClonalOrigin_pair
import clonal_genealogy
import matplotlib.pyplot as plt
```

```python
tree = ClonalTree(n=15)
```

```python
rho_site = 0.2
L = 100000
delta = 300
k = 2000
theta_site = 0.2
```

```python
ARG = ClonalOrigin_pair.ARG(tree, rho_site, L, delta, k)
```

```python
ARG.node_site
```

```python
ClonalOrigin_pair_time = []
add_mutation_time = []

for i in range(1000):
    np.random.seed(i)

    start_time_1 = time.time()
    ARG = ClonalOrigin_pair.ARG(tree, rho_site, L, delta, k)
    end_time_1 = time.time()
    ClonalOrigin_pair_time.append(end_time_1 - start_time_1)

    start_time_2 = time.time()
    ARG.add_mutation(theta_site)
    end_time_2 = time.time()
    add_mutation_time.append(end_time_2 - start_time_2)

    if i % 100 == 0:
        print(f"Complete {i} iterations.")
```

```python
np.mean(ClonalOrigin_pair_time), np.mean(add_mutation_time)
```
