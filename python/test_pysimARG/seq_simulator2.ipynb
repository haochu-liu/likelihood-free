## Try and test simulator

```python
rho_site = 0.02
theta_site = 0.05
L = 2000
delta = 300
```

```python
ARG_sim = ARG(tree1, rho_site, L, delta, L, "seq")
```

```python
ARG_sim.edge.shape, ARG_sim.node_mat.shape
```

```python
node_site = add_mutation(ARG_sim, theta_site)
node_site.shape
```

```python
node_site
```

```python
mat = node_site[:15, :]
mat
```

```python
s = 0
s_idx = []
for i in range(2000):
    s_false = np.sum(mat[:, i] == False)
    s_true = np.sum(mat[:, i] == True)
    if s_false > 0 and s_true > 0:
        s += 1
        s_idx.append(i)
s, s_idx
```

```python
has_true = mat.any(axis=0)
has_false = ~mat.all(axis=0)
count_S = (has_true & has_false).sum()
count_S
```

```python
has_seg = has_true & has_false
has_seg.sum()
```

```python
np.sum(np.where(has_seg)[0] == np.array(s_idx))
```

```python
idx_seg = np.where(has_seg)[0]
idx_seg
```

```python
s = 0
for i in range(idx_seg[:10].size - 1):
    for j in range(i + 1, idx_seg[:10].size):
        print(i, j)
        dist_ij = idx_seg[j] - idx_seg[i]
        print(dist_ij)
        s += 1
s
```

```python
ld_near, ld_far, g3_near, g3_far = 0, 0, 0, 0
if idx_seg.size >= 2:
    for i in range(idx_seg.size - 1):
        for j in range(i + 1, idx_seg.size):
            dist_ij = idx_seg[j] - idx_seg[i]
            idx_pair = [idx_seg[i], idx_seg[j]]
            if dist_ij < L/2:
                ld_near += LD_r(mat[:, idx_pair])
                g3_near += G3_test(mat[:, idx_pair])
            else:
                ld_far += LD_r(mat[:, idx_pair])
                g3_far += G3_test(mat[:, idx_pair])
    print([ld_near, ld_far, g3_near, g3_far])
else:
    print(0)
```

```python
ld_near, ld_far, g3_near, g3_far = 0, 0, 0, 0
for i in range(2000-1):
    for j in range(i + 1, 2000):
        dist_ij = j - i
        idx_pair = [i, j]
        if dist_ij < L/2:
            ld_near += LD_r(mat[:, idx_pair])
            g3_near += G3_test(mat[:, idx_pair])
        else:
            ld_far += LD_r(mat[:, idx_pair])
            g3_far += G3_test(mat[:, idx_pair])
print([ld_near, ld_far, g3_near, g3_far])
```
