```python
import numpy as np
import sys
sys.path.append('../../pysimARG')
import json
from G3_test import G3_test
from LD_r import LD_r
```

```python
with open('../../pysimARG/test_data/summary_stats.json', 'r') as f:
    summary_stats = json.load(f)

# Verify the type and content
print(type(summary_stats))  # <class 'dict'>
print(summary_stats)
```

```python
r_list = summary_stats['r']
g3_list = summary_stats['g3']
leaf_mat = np.array(summary_stats['leaf'])
```

```python
for i in range(len(r_list)):
    test_mat = leaf_mat[:, i*2:(i+1)*2]
    test_r = LD_r(test_mat)
    test_g3 = G3_test(test_mat)
    print(f"Summary stats for dataset {i+1}:")
    print(f"Consistency check for r: {test_r} vs {r_list[i]}")
    print(f"Consistency check for g3: {test_g3} vs {g3_list[i]}")
    print("-" * 50)
```
