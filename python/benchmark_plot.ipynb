# Plots for the benchmarking test

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

## Load data

```python
model_names = ['NLE', 'NPE', 'NRE']
sim_budgets = ['200', '500', '1000', '2000', '5000', '10000', '50000']
```

```python
nle_coverage = np.loadtxt('../data/benchmark/nle_coverage_results.csv', delimiter=',')
npe_coverage = np.loadtxt('../data/benchmark/npe_coverage_results.csv', delimiter=',')
nre_coverage = np.loadtxt('../data/benchmark/nre_coverage_results.csv', delimiter=',')
coverage_results = [nle_coverage, npe_coverage, nre_coverage]
```

```python
nle_D = np.loadtxt('../data/benchmark/nle_sbc_D.csv', delimiter=',')
npe_D = np.loadtxt('../data/benchmark/npe_sbc_D.csv', delimiter=',')
nre_D = np.loadtxt('../data/benchmark/nre_sbc_D.csv', delimiter=',')
D_results = [nle_D, npe_D, nre_D]
```

```python
nle_p = np.loadtxt('../data/benchmark/nle_sbc_p_values.csv', delimiter=',')
npe_p = np.loadtxt('../data/benchmark/npe_sbc_p_values.csv', delimiter=',')
nre_p = np.loadtxt('../data/benchmark/nre_sbc_p_values.csv', delimiter=',')
p_results = [nle_p, npe_p, nre_p]
```

```python
nle_error = np.load('../data/benchmark/nle_mean_error_results.npy')
npe_error = np.load('../data/benchmark/npe_mean_error_results.npy')
nre_error = np.load('../data/benchmark/nre_mean_error_results.npy')
mean_error_results = [nle_error, npe_error, nre_error]
```

## Figures

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Joint Coverage', fontsize=14, fontweight='bold', y=1.05)

results = [coverage_results[i][0] for i in range(len(model_names))]
for i, ax in enumerate(axes):
    ax.plot(sim_budgets, results[i], marker='o', linestyle='-', color='#1f77b4', linewidth=1.5)
    ax.hlines(y=0.95, xmin=sim_budgets[0], xmax=sim_budgets[-1], color='red', linestyle='--', linewidth=1, label='Target coverage: 95.0%')
    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(sim_budgets)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    
    ax.tick_params(axis='y', length=0)
    ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1.0))

axes[0].set_ylabel('Coverage', fontsize=12)
axes[0].legend(loc='lower right', fontsize=10, frameon=False)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.ylim(0, 1)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_coverage.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: SBC Test P-Values', fontsize=14, fontweight='bold', y=1.05)

results_rho = [p_results[i][0] for i in range(len(model_names))]
results_theta = [p_results[i][1] for i in range(len(model_names))]
for i, ax in enumerate(axes):
    ax.plot(sim_budgets, results_rho[i], marker='o', linestyle='-', color='#1f77b4', linewidth=1.5, label=r'Marginal posterior $\rho_s$')
    ax.plot(sim_budgets, results_theta[i], marker='o', linestyle='-', color='#ff7f0e', linewidth=1.5, label=r'Marginal posterior $\theta_s$')
    ax.hlines(y=0.05, xmin=sim_budgets[0], xmax=sim_budgets[-1], color='red', linestyle='--', linewidth=1, label='Significance level (0.05)')
    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(sim_budgets)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(axis='y', length=0)

axes[0].set_ylabel('p-value', fontsize=12)
axes[0].legend(loc='upper left', fontsize=10, frameon=False)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.ylim(0, 1)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_sbc_p.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: SBC Test P-Values', fontsize=14, fontweight='bold', y=1.05)

results_rho = [p_results[i][0] for i in range(len(model_names))]
results_theta = [p_results[i][1] for i in range(len(model_names))]
for i, ax in enumerate(axes):
    ax.plot(sim_budgets, results_rho[i], marker='o', linestyle='-', color='#1f77b4', linewidth=1.5, label=r'Marginal posterior $\rho_s$')
    ax.plot(sim_budgets, results_theta[i], marker='o', linestyle='-', color='#ff7f0e', linewidth=1.5, label=r'Marginal posterior $\theta_s$')
    ax.hlines(y=0.05, xmin=sim_budgets[0], xmax=sim_budgets[-1], color='red', linestyle='--', linewidth=1, label='Significance level (0.05)')
    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(sim_budgets)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    
    ax.tick_params(axis='y', length=0)
    ax.set_yscale('log')

axes[0].set_ylabel('p-value', fontsize=12)
axes[0].legend(loc='lower right', fontsize=10, frameon=False)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_sbc_logp.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Parameter Estimation Error', fontsize=14, fontweight='bold', y=1.05)

raw_results = [mean_error_results[i][:, :, 0] for i in range(len(model_names))]
for i, ax in enumerate(axes):
    means = np.mean(raw_results[i], axis=0)
    stds = np.std(raw_results[i], axis=0)

    ax.errorbar(
        sim_budgets, means, yerr=stds, 
        fmt='-o', color='#1f77b4', 
        linewidth=1.5, elinewidth=1.5, 
        capsize=0,
        markersize=6
    )

    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(sim_budgets)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(axis='y', length=0)

axes[0].set_ylabel(r'Estimation Error ($L_2$)', fontsize=12)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_error_mean.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Parameter Estimation Error', fontsize=14, fontweight='bold', y=1.05)

raw_results = [mean_error_results[i][:, :, 0] for i in range(len(model_names))]
for i, ax in enumerate(axes):
    
    median = np.median(raw_results[i], axis=0)
    lower_bounds = np.percentile(raw_results[i], 25, axis=0)
    upper_bounds = np.percentile(raw_results[i], 75, axis=0)
    lower_errors = median - lower_bounds
    upper_errors = upper_bounds - median
    yerr = [lower_errors, upper_errors]

    ax.errorbar(
        sim_budgets, median, yerr=yerr, 
        fmt='-o', color='#1f77b4', 
        linewidth=1.5, elinewidth=1.5, 
        capsize=4,
        markersize=6
    )

    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(sim_budgets)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(axis='y', length=0)

axes[0].set_ylabel(r'Estimation Error ($L_2$)', fontsize=12)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_error_median.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
fig.suptitle('Benchmark Performance: Parameter Estimation Error', fontsize=14, fontweight='bold', y=1.05)

raw_results = [mean_error_results[i][:, :, 0] for i in range(len(model_names))]
x_positions = np.arange(len(sim_budgets))

for i, ax in enumerate(axes):
    bplot = ax.boxplot(
        raw_results[i],
        positions=x_positions,
        patch_artist=True,
        widths=0.4,
        showfliers=False
    )
    
    for patch in bplot['boxes']:
        patch.set_facecolor('white')
        patch.set_edgecolor('#1f77b4')
        patch.set_linewidth(1.5)
        
    for element in ['whiskers', 'caps', 'medians']:
        for line in bplot[element]:
            line.set_color('#1f77b4')
            line.set_linewidth(1.5)

    ax.set_title(model_names[i], fontsize=12, pad=10)
    ax.set_xticks(x_positions)
    ax.set_xticklabels(sim_budgets, rotation=90)

    ax.grid(axis='y', linestyle='-', color='lightgray', alpha=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('lightgray')
    ax.tick_params(axis='y', length=0)

axes[0].set_ylabel(r'Estimation Error ($L_2$)', fontsize=12)
fig.text(0.5, -0.05, 'Number of Simulations', ha='center', va='center', fontsize=12)
plt.subplots_adjust(wspace=0.05, bottom=0.2)
plt.savefig('../output/benchmark_error_box.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
